Write a Python program that:
• Asks the user to enter 10 numbers.
• Skips negative numbers using continue.
• Stops taking input if the user enters 0 using break.
• Adds only even positive numbers using an augmented assignment operator.
• Prints the final sum.

```python
total = 0

for i in range(1, 11):
   num = int(input(f"Enter number {i}: "))

   if num == 0:
       print("Input stopped by user.")
       break
   if num < 0:
       continue
   if num % 2 == 0:
       total += num

print("Sum of positive even numbers:", total)
```

Write a Python program that:
• Asks the user to enter a number.
• Prints its multiplication table from 1 to 12 using a forloop.
• Skips printing multiples of 3 using continue.
• Stops the table if the product is greater than 50 using break.


```python
num = int(input("Enter a number: "))

for i in range(1, 13):
   product = num * i
   if product > 50:
       break
   if product % 3 == 0:
       continue
   print(f"{num} x {i} = {product}")
```

3) Write a Python program that:
• Chooses a secret number between 1 and 20.
• Asks the user to guess the number.
• Uses a while loop to allow unlimited attempts.
• Stops the loop if the user guesses correctly (break).
• Skips invalid guesses (less than 1 or greater than 20) using continue.
• Keeps track of the number of valid attempts using an augmented assignment (+=).

```python
secret_number = 7
attempts = 0

while True:
   guess = int(input("Guess the number (1-20): "))

   if guess < 1 or guess > 20:
       print("Invalid guess, try again.")
       continue

   attempts += 1

   if guess == secret_number:
       print(f"Congratulations! You guessed it in {attempts} attempts.")
       break
   elif guess < secret_number:
       print("Too low.")
   else:
       print("Too high.")
```

4) Write a Python program that:
• Asks the user to enter a number n.
• Counts how many numbers from 1 to n are divisible by 4.
• Stops counting if a number greater than 50 is found (break).
• Skips numbers divisible by 8 using continue.
• Prints the count at the end.


```python
n = int(input("Enter a number: "))
count = 0

for i in range(1, n + 1):
   if i > 50:
       break
   if i % 8 == 0:
       continue
   if i % 4 == 0:
       count += 1
print("Count of numbers divisible by 4 (excluding multiples of 8, stopping at >50):", count)
```


5) Write a Python program that:
• Prints a triangle pattern of numbers.
• For i from 1 to 5, print numbers from 1 to i.
• Skip printing the number 3 in the triangle using continue.
• Stop printing the triangle if i * j > 6 using break.


```python
for i in range(1, 6):
   for j in range(1, i + 1):
       if j == 3:
           continue
       if i * j > 6:
           break
       print(j, end=" ")
   print()
```
