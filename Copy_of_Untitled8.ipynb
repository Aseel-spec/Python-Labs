 Food Order Manager
 This program allows user to add, display, and search food orders

```python
orders = []

def add_order():
    food_name = input("Enter food name: ")
    quantity = int(input("Enter quantity: "))
    price = float(input("Enter price: "))

    order = {
        "food_name": food_name,
        "quantity": quantity,
        "price": price
    }

    orders.append(order)
    print("Order added successfully!\n")


def display_orders():
    if len(orders) == 0:
        print("No orders found.\n")
        return

    print("\nFood Orders:")
    for i, order in enumerate(orders, start=1):
        print(f"{i}. Name: {order['food_name']}, Quantity: {order['quantity']}, Price: {order['price']}")
    print()


def search_order():
    search_name = input("Enter food name to search: ")

    found = False
    for order in orders:
        if order["food_name"].lower() == search_name.lower():
            print("Order Found:")
            print(f"Name: {order['food_name']}, Quantity: {order['quantity']}, Price: {order['price']}\n")
            found = True
            break

    if not found:
        print("Order not found.\n")


def main():
    while True:
        print("===== Food Order Manager =====")
        print("1. Add Order")
        print("2. Display Orders")
        print("3. Search Order")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_order()
        elif choice == "2":
            display_orders()
        elif choice == "3":
            search_order()
        elif choice == "4":
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Try again.\n")


main()
```
